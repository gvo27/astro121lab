## 5.2 Part A
We first collect data with the default anti-aliasing filter.

```python
# ugradio.sdr.SDR??
```

```python
# range of samp frequencies
samp_freqs = np.arange(1000, 2400, 200)

# array containing the voltage outputs for all the different frequencies
all_data = np.array([])
for freq in samp_freqs:
    fname = 'freq' + str(freq) + '.npy'
    data = np.load(fname)
    all_data = np.append(all_data, data)
all_data = all_data.reshape(7, 2048)
```

```python
# Note: need to convert x axis to time using frequency to time conversion

f = 1000 # kHZ
t = 1/f # ms
```

```python
for i in np.arange(len(all_data)-1):
    freq = samp_freqs[i]
    data = all_data[i]
    plt.figure()
    plt.plot(np.arange(len(data)), data)
    plt.scatter(np.arange(len(data)), data)
    plt.xlim(0, 20)
    plt.title('Freq = ' + str(freq) + ' kHz')
    plt.xlabel('Time [s]')
    plt.ylabel('Voltage [V]')
    plt.show()
```
